```python
import pandas as pd
import numpy as np

# Configuración
# -----------------------------------------------------------------------
pd.set_option('display.max_columns', None) # para poder visualizar todas las columnas de los DataFrames
pd.set_option('display.max_colwidth', None)
```

```python
df = pd.read_csv('gym_members_exercise_tracking.csv')
```

```python
df.head(20)
```

```python
# una vez que tengamos nuestro DataFrame preparado con todas las columnas que queremos vamos a crear una función que no haga una exploración inicial del conjunto de datos
def exploracion_dataframe(dataframe):
    
    print(f"Los duplicados que tenemos en el conjunto de datos son: {dataframe.duplicated().sum()}")
    print("\n ..................... \n")
    
    
    # generamos un DataFrame para los valores nulos
    print("Los nulos que tenemos en el conjunto de datos son:")
    
    df_nulos_num = pd.DataFrame(dataframe.isnull().sum() / dataframe.shape[0] * 100, columns = ["%_nulos"])
    display(df_nulos_num[df_nulos_num["%_nulos"] > 0])

    # Nulos en columnas categóricas
    print("Los nulos que tenemos en el conjunto de datos (Categóricas) son:")
    df_nulos_cat = pd.DataFrame(dataframe.select_dtypes(include="object").isnull().sum() / dataframe.shape[0] * 100, columns=["%_nulos"])
    display(df_nulos_cat[df_nulos_cat["%_nulos"] > 0])

    print("\n ..................... \n")
    print(f"Los tipos de las columnas son:")
    display(pd.DataFrame(dataframe.dtypes, columns = ["tipo_dato"]))
    
    
    print("\n ..................... \n")
    print("Los valores que tenemos para las columnas númericas son: ")
    dataframe_numericas = dataframe.select_dtypes(include = np.number)
    
    for col in dataframe_numericas.columns:
        print(f"La columna {col.upper()} tiene las siguientes valores únicos:")
        display(pd.DataFrame(dataframe[col].value_counts()).head())   


    print("\n ..................... \n")
    print("Los valores que tenemos para las columnas categóricas son: ")
    dataframe_categoricas = dataframe.select_dtypes(include = "O") 

    for col in dataframe_categoricas.columns:
        print(f"La columna {col.upper()} tiene las siguientes valore únicos:")
        display(pd.DataFrame(dataframe[col].value_counts()).head())    
    
   
```

```python
exploracion_dataframe(df)
```
